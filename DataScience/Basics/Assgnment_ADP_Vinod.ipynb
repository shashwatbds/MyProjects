## **Understanding the Data**
We are given two csv files namely **input_features** and **target_values**. 

**Input_features contains the below information :**

**1) Building ID:** Primary key to identify each each building uniquely.  
**2) Location Details:**
Features describing the geographic region in which building exists
> 

*   **geo_level_1_id** - Largest level of Region

> **Data Type:** *Int*

> **Possible Values:** *0-30*


*   **geo_level_2_id** - Sub-region under geo_level_1_id.

> **Data Type:** *Int*

> **Possible Values:** *0-1427*


*   **geo_level_3_id**- Sub-region under geo_level_2_id. Smallest level of sub-region.

> **Data Type:** *Int*

> **Possible Values:**  *0-12567*


**3) foundation_type:** - Categorical feature describing type of foundation used in building.

> **Data Type:** *object(Categorical)*

> **Possible Values:**  *h,i,r,u,w*


**4) count_floors_pre_eq:** - Number of floors in th ebuilding before earthquake.

> **Data Type:** *int*

**5) area_percentage:** - Normalized area of thebuilding footprint.

> **Data Type:** *Int*


**6) height_percentage:** - Normalized height ofthe building footprint.

> **Data Type:** *Int*

**7) land_surface_condition:** - Normalized height ofthe building footprint.

> **Data Type:** *object(Categorical)*

> **Possible Values:**  *n, o, t*


**8) roof_type:** - Type of roof used in the building.

> **Data Type:** *object(Categorical)*

> **Possible Values:**  *n, q, x*

**9) ground_floor_type:** - Type of the ground floor.

> **Data Type:** *object(Categorical)*

> **Possible Values:**  *f, m, v, x, z*


**10) other_floor_type:** - Type of construction used in floors other than ground floor.

> **Data Type:** *object(Categorical)*

> **Possible Values:**  *j, q, s, x*


**11) position:** - Position of the building.

> **Data Type:** *object(Categorical)*

> **Possible Values:**  *o, s, t*


**12) plan_configuration:** - Plan configuration of the building.

> **Data Type:** *object(Categorical)*

> **Possible Values:**  *a, c, d, f, m, n, o, q, s, u*

**13) Super Structure Type:** - Various binary columns describing the type of super strucutre use per building.

> **Data Type:** *int(Binary)*

> **Possible Values:**  *0,1*

> **Super Structure Types:** *adobe_mud, mud_mortar_stone, stone, cement_mortar_stone, mud_mortar_brick, cement_mortar_brick, brick, bamboo, rc_non_engineered, rc_engineered, other*


**14) legal_ownership_status:** - Legal ownership status of land where the building was made.

> **Data Type:** *object(Categorical)*

> **Possible Values:**  *a, r, v, w*

**15) count_families:** - Number of families living in the building.

> **Data Type:** *Int*


**16) has_secondary_use:** - Binary property indicating if the building was used for a secondary perpose or not.

> **Data Type:** *int(Binary)*

> **Possible Values:**  *0,1*

**17) Secondary use kind of building:** - Binary properties that indicating the kind of secondary use building had.

> **Data Type:** *int(Binary)*

> **Possible Values:**  *0,1*

> **Secondary Use Types:** *agriculture, hotel, rental, institution, school, industry, health post, government office, used by police, other use*


**Target_values contains the below information :**

**1) Building ID:** Primary key to identify each each building uniquely.  
**2) Damage Grade:** Damage grade on the building after earthquake. 

> **Data Type:** *int(Category)*

> **Possible Values:** *1,2,3*

> **Damage Grade 1:** *Low Damage*

> **Damage Grade 1:** *Medium amount of damage*

```python
from matplotlib import colors
## Frequency histograms
## Regions 
##  1) geo_level_1_id
fig, axs = plt.subplots(1,3,tight_layout=True, figsize=(20,5))
fig.suptitle("Frequency Damage across region", fontsize=25, y=1.10 ) # y value to place sup title slightly outside the figure.

# N is the count in each bin, bins is the lower-limit of the bin
N, bins, patches = axs[0].hist(buildings_df['geo_level_1_id'], bins='auto')
axs[0].set_xlabel("geo_level_1_id")
axs[0].set_ylabel("Frequency")

# We'll color code by height, but you could use any scalar
fracs = N / N.max()
norm = colors.Normalize(fracs.min(), fracs.max())
# Now, we'll loop through our objects and set the color of each accordingly
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)

##  2) geo_level_2_id
N, bins, patches = axs[1].hist(buildings_df['geo_level_2_id'], bins=70)
axs[1].set_xlabel("geo_level_2_id")

##  3) geo_level_3_id
N, bins, patches = axs[2].hist(buildings_df['geo_level_3_id'], bins='auto')
axs[2].set_xlabel("geo_level_3_id")

fig.subplots_adjust(top=2)

fig.show()

## Observations
## Geo Level 1: Freequency of damage caused in id 7, 11, 18 and 27.
## Geo Level 2: A couple of obvious high intensity areas identified, can be looked upon in greater depth during EDA. 
## Geo Level 3: Too many possible values to identify any pattern here by just using histogram.

## Inferences:
## 1) Since geo level 1 is the largest level of region and we see some obvious highly impact regions.
## 2) With geo level ids with highest damage, we can analyze which region has highest propotion buildings with highest and lowest damage grade respectively.
```

```python
## Which Geo_level_1_id has maximum buildings with damage grade 3
pd.crosstab(buildings_df.geo_level_1_id, buildings_df.damage_grade).plot(kind='bar', figsize=(20,7))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Damage Grade distribution per geo_level_1_id", fontsize=25, y=1.10)

plt.show()

## Observation
## 1. Geo_level_1_id=17 seems to have highest number of buildings with damage grade = 3 (Complete Destruction)
```
