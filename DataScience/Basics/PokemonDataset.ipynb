From the Pokémon dataset

a) How many Pokémon of each type are there?
 
b) Which Pokémon type is the most powerful?

c) Find average values, variance for each property and plot them (e.g. bar chart)

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
pokemon_df = pd.read_csv("https://raw.githubusercontent.com/shashwatbds/MyProjects/main/DataScience/Dataset/all_pokemon.csv")
```

```python
pokemon_df.head()
```

```python
# How many Pokémon of each type are there?
pokemon_df.Type.unique().size
```

```python
# Which Pokémon type is the most powerful?
pokemon_df[pokemon_df.Total == pokemon_df.Total.max()]['Type']
```

```python
pokemon_df.describe()
```

```python
# Find average values, variance for each property and plot them (e.g. bar chart)
#mean = pokemon_df.describe().loc['mean']
mean_series = pd.Series(pokemon_df.mean(), name="Mean")
var_series = pd.Series(pokemon_df.var(), name="Varience")
describe_df = pd.concat([mean_series, var_series], axis=1)
describe_df[1:]
```

```python
fig, axs = plt.subplots(2,3,tight_layout=True, figsize=(20,10))

fig.suptitle("Pokemon Histogram Plots", fontsize=25, y=1.10, fontweight='bold' ) 

axs[0,0].hist(pokemon_df['HP']) # plot histogram for passed column name
axs[0,0].set_xlabel("HP", fontsize = 'large', fontweight = 'bold') # Label for x-axis
axs[0,0].set_ylabel("Frequency", fontsize = 'large', fontweight = 'bold') # Label for y-axis

axs[0,1].hist(pokemon_df['Attack']) # plot histogram for passed column name
axs[0,1].set_xlabel("Attack", fontsize = 'large', fontweight = 'bold') # Label for x-axis

axs[0,2].hist(pokemon_df['Defense']) # plot histogram for passed column name
axs[0,2].set_xlabel("Defense", fontsize = 'large', fontweight = 'bold') # Label for x-axis

axs[1,0].hist(pokemon_df['Sp. Atk']) # plot histogram for passed column name
axs[1,0].set_xlabel("Special Attack", fontsize = 'large', fontweight = 'bold') # Label for x-axis
axs[1,0].set_ylabel("Frequency", fontsize = 'large', fontweight = 'bold') # Label for y-axis

axs[1,1].hist(pokemon_df['Sp. Def']) # plot histogram for passed column name
axs[1,1].set_xlabel("Special Defense", fontsize = 'large', fontweight = 'bold') # Label for x-axis

axs[1,2].hist(pokemon_df['Speed']) # plot histogram for passed column name
axs[1,2].set_xlabel("Speed", fontsize = 'large', fontweight = 'bold') # Label for x-axis

plt.show() # Display the dashboard.
```
